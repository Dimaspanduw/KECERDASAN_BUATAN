```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('motorbike.csv') #untuk membaca dataset 
data.head()
```

1. GRAFIK BATANG

```python
#Mengelompokkan data berdasarkan Version dan menghitung jumlah datanya
data_version = data['version'].value_counts().sort_index()

# Membuat grafik batang
plt.figure(figsize=(20, 6))
sns.barplot(x=data_version, y=data_version.values, palette='viridis')
plt.title('Data total versi pasar sepeda motor')
plt.xlabel('Jumlah version')
plt.ylabel('Jumlah Data')

# Menampilkan grafik batang
plt.show()
```

2. HISTOGRAM

```python
sns.histplot(data['mileage'], bins=15, kde=True, color='skyblue')
plt.title('Data total jarak tempuh pasar sepeda motor')
plt.xlabel('mileage')
plt.ylabel('Frequency')
plt.show()
```

3. PIE CHAR

```python
# Selecting top 10 price
top_price = data['price'].value_counts().head(10)

# Plotting the pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_price, labels=top_price.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Data total Harga pasar sepeda motor')
plt.show()
```

4. HEATMAP

```python
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()

# Plotting the heatmap
plt.figure(figsize=(22, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation motorbike marketplace')
plt.xlabel('Numerical Features')
plt.ylabel('Numerical Features')
plt.show()
```
